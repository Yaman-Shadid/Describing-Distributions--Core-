```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Define the file path
file_path = "G:\\Github Desktop Reps\\Coding Dojo\\Describing-Distributions--Core-\\Data\\Medical_Dataset.csv"

# Load the dataset into a DataFrame named 'df'
df = pd.read_csv(file_path)

# Check the first few rows to ensure it's loaded correctly
df.head()
```

```python
# Define Features
features = ['VitD_levels', 'Doc_visits', 'TotalCharge']

for feature in features:
    plt.figure(figsize=(10, 6))
    
    # Plot histogram with KDE
    sns.histplot(df[feature], kde=True)
    
    # Mean and Median
    mean = df[feature].mean()
    median = df[feature].median()
    
    # Standard Deviation
    std = df[feature].std()
    
    # Plotting lines for mean, median, and std deviations
    plt.axvline(mean, color='red', linestyle='--', label='Mean')
    plt.axvline(median, color='green', linestyle='-', label='Median')
    plt.axvline(mean + std, color='black', linestyle='-', label='+1 std deviation')
    plt.axvline(mean - std, color='black', linestyle='-', label='-1 std deviation')
    
    # Highlighting the range between +1 and -1 std
    plt.axvspan(mean - std, mean + std, color='yellow', alpha=0.2)
    
    plt.title(f"Histogram for {feature}")
    plt.legend()
    plt.show()
```
